## Histogram of Male and Female Weights¶
In this step, histograms are created to visualize the distribution of body weights for female and male participants. Two subplots are drawn on a single figure using "matplotlib.pyplot.subplot".

The top subplot displays the histogram of female weights, while the bottom subplot shows the histogram of male weights. The x-axis limits are set to be identical for both plots using "matplotlib.pyplot.xlim" so that the distributions can be compared fairly.

```python
import matplotlib.pyplot as plt
import numpy as np

female_weights = female[:,0]
male_weights = male[:,0]

female_weights = female_weights[~np.isnan(female_weights)]
male_weights = male_weights[~np.isnan(male_weights)]

xmin = min(np.min(female_weights), np.min(male_weights))
xmax = max(np.max(female_weights), np.max(male_weights))

plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.hist(female_weights, bins=20)
plt.title("Female Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.xlim(xmin, xmax)

plt.subplot(2,1,2)
plt.hist(male_weights, bins=20)
plt.title("Male Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.xlim(xmin, xmax)

plt.tight_layout()
plt.show()
```

## Description
The histograms display the distribution of body weights for female and male participants. By setting identical x-axis limits for both subplots, it becomes easier to visually compare the two distributions.

The male weight distribution generally appears shifted toward higher values compared to the female distribution, indicating that male participants tend to have higher body weights. Additionally, the spread of male weights appears slightly larger, suggesting greater variability in male body weight.

## Numerical Aggregates of Male and Female Weights

In this step, basic numerical aggregates of the weight distributions for male and female participants are computed. These statistics help summarize the data in terms of location, dispersion, and shape.
Measures of location:
- Mean
- Median
Measures of dispersion:
- Standard deviation
- Variance
Measures of shape:
- Skewness
- Kurtosis
These statistics help us understand the central tendency, variability, and overall distribution pattern of the body weights.


```python
import numpy as np

female_weights = female[:,0]
male_weights = male[:,0]

female_weights = female_weights[~np.isnan(female_weights)]
male_weights = male_weights[~np.isnan(male_weights)]

# ---- Female statistics ----
female_mean = np.mean(female_weights)
female_median = np.median(female_weights)
female_std = np.std(female_weights)
female_var = np.var(female_weights)

# ---- Male statistics ----
male_mean = np.mean(male_weights)
male_median = np.median(male_weights)
male_std = np.std(male_weights)
male_var = np.var(male_weights)

female_skew = np.mean(((female_weights - female_mean)/female_std)**3)
male_skew = np.mean(((male_weights - male_mean)/male_std)**3)

female_kurt = np.mean(((female_weights - female_mean)/female_std)**4)
male_kurt = np.mean(((male_weights - male_mean)/male_std)**4)

print("Female Weights Statistics")
print("Mean:", female_mean)
print("Median:", female_median)
print("Standard Deviation:", female_std)
print("Variance:", female_var)
print("Skewness:", female_skew)
print("Kurtosis:", female_kurt)

print("\nMale Weights Statistics")
print("Mean:", male_mean)
print("Median:", male_median)
print("Standard Deviation:", male_std)
print("Variance:", male_var)
print("Skewness:", male_skew)
print("Kurtosis:", male_kurt)
```


## Description
The computed numerical aggregates provide a summary of the weight distributions for male and female participants.

The mean and median represent the central tendency of the distributions. In general, the average weight of male participants is higher than that of female participants.

The standard deviation and variance measure the dispersion of the data. The male weight distribution typically shows a larger standard deviation, indicating that male weights are more spread out compared to female weights.

The skewness value indicates the shape of the distribution. A positive skewness suggests that the distribution is slightly right-skewed, meaning there are some higher weight values that extend the right tail of the distribution.

Kurtosis provides information about the peakedness of the distribution. Higher kurtosis values indicate heavier tails or more extreme values.

Overall, the male weight distribution tends to have a higher central value and greater variability than the female weight distribution.


## Adding BMI Column to the Female Dataset

In this step, the Body Mass Index (BMI) of each female participant is calculated and added as the eighth column to the female matrix.

BMI is a commonly used indicator of body fat based on weight and height. It is calculated using the following formula:

BMI = weight / (height²)

Since the height values in the dataset are given in centimeters, they must first be converted into meters before computing BMI.


```python

female_weights = female[:,0]
female_height_cm = female[:,1]

female_height_m = female_height_cm / 100

# Calculate BMI
female_bmi = female_weights / (female_height_m ** 2)

# Add BMI as the 8th column
female = np.column_stack((female, female_bmi))
print("New female matrix shape:", female.shape)
print(female[:5])
```

## Description
The Body Mass Index (BMI) values were successfully calculated for all female participants using their weight and height measurements. The height values were first converted from centimeters to meters to ensure correct computation of BMI.

The calculated BMI values were then added as the eighth column of the female matrix. As a result, the dataset now contains an additional variable that represents the relative body mass of each participant. This new variable can be used for further statistical analysis and comparison with other body measurements.


## Standardizing the Female Dataset (Z-Scores)

In this step, a new matrix called `"zfemale" is created by standardizing all columns of the female dataset. Standardization transforms the values of each variable so that they have a mean of 0 and a standard deviation of 1.
This process is useful because the body measurements in the dataset are expressed in different units and ranges. By converting them into z-scores, the variables become comparable on the same scale.
The z-score of a value is calculated using the formula:
z = (x − mean) / standard deviation
Where:
- x is the original value
- mean is the average of the column
- standard deviation measures the spread of the data


```python
# Compute column means ignoring NaN
means = np.nanmean(female, axis=0)

# Compute column standard deviations ignoring NaN
stds = np.nanstd(female, axis=0)

# Standardize dataset (z-score)
zfemale = (female - means) / stds
print("First 5 rows of standardized female matrix:")
print(zfemale[:5])
```

### Description

The female dataset was standardized using the z-score transformation. Since the dataset contained some missing values, NumPy functions that ignore NaN values were used when computing the mean and standard deviation. 

After standardization, each column in the matrix has an approximate mean of 0 and a standard deviation of 1. This transformation allows variables measured in different units to be compared on the same scale.

## Computing Waist-to-Height Ratio and Waist-to-Hip Ratio

In this step, two additional body measurement ratios are computed for both male and female participants.

1. Waist-to-Height Ratio (WHtR)  
   This ratio is calculated by dividing waist circumference by height.

2. Waist-to-Hip Ratio (WHR)  
   This ratio is calculated by dividing waist circumference by hip circumference.

These ratios are commonly used indicators of body fat distribution and health risk. After computing the ratios, they are added as two new columns to both the male and female matrices.


```python
male_height = male[:,1]
male_hip = male[:,5]
male_waist = male[:,6]

male_wthr = male_waist / male_height
male_whr = male_waist / male_hip

male = np.column_stack((male, male_wthr, male_whr))


female_height = female[:,1]
female_hip = female[:,5]
female_waist = female[:,6]

female_wthr = female_waist / female_height
female_whr = female_waist / female_hip

female = np.column_stack((female, female_wthr, female_whr))

print("Male matrix shape:", male.shape)
print("Female matrix shape:", female.shape)

print("\nFirst 5 rows of male matrix:")
print(male[:5])

print("\nFirst 5 rows of female matrix:")
print(female[:5])
```

### Discussion

The waist-to-height ratio and waist-to-hip ratio were successfully calculated for both male and female participants. These ratios provide useful indicators of body fat distribution.

The waist-to-height ratio helps identify central obesity by comparing waist circumference with overall body height. Similarly, the waist-to-hip ratio measures the proportion of fat stored around the waist relative to the hips.

After computing these values, the ratios were added as two additional columns to both the male and female matrices. As a result, each dataset now contains two new variables that can be used for further analysis and comparison between male and female participants.

## Advantages and Disadvantages of BMI, Waist-to-Height Ratio, and Waist-to-Hip Ratio
### Body Mass Index (BMI)
**Advantages**
- BMI is very easy to calculate because it only requires weight and height measurements.
- It is widely used in medical and public health research to classify individuals as underweight, normal weight, overweight, or obese.
- BMI provides a quick estimate of overall body mass and can be applied to large populations.
**Disadvantages**
- BMI does not distinguish between muscle mass and body fat. For example, athletes with high muscle mass may have a high BMI even though they are healthy.
- It does not provide information about the distribution of body fat in the body.
- BMI may not accurately represent health risk for all individuals.
---
### Waist-to-Height Ratio (WHtR)
**Advantages**
- The waist-to-height ratio is a good indicator of central or abdominal obesity.
- It is considered a useful predictor of cardiovascular disease risk.
- The calculation is simple and requires only waist circumference and height measurements.

**Disadvantages**
- The ratio depends on accurate waist measurements, which may sometimes be inconsistent.
- It may vary across different populations and age groups.
- It does not consider other body composition factors such as muscle mass.
---
### Waist-to-Hip Ratio (WHR)
**Advantages**
- Waist-to-hip ratio helps identify patterns of fat distribution in the body.
- It is useful in assessing the risk of heart disease and metabolic disorders.
- It provides more information about fat distribution than BMI alone.
**Disadvantages**
- Accurate measurement of waist and hip circumference can be difficult.
- It may not reflect overall body fat percentage.
- The ratio may vary due to body shape differences rather than health risk alone.
---
### Summary
BMI provides a simple measure of overall body mass, while waist-to-height and waist-to-hip ratios provide additional information about fat distribution in the body. Using these measures together can give a better understanding of an individual's health risk than relying on a single measurement alone.

## Identifying Participants with Lowest and Highest BMI

In this step, the standardized body measurements of female participants with the lowest and highest Body Mass Index (BMI) values are identified. 

The BMI values are sorted using the NumPy function 'numpy.argsort', which returns the indices that would sort the BMI values. Using these indices, the standardized measurements of the five participants with the lowest BMI and the five participants with the highest BMI are selected from the standardized matrix 'zfemale'.

This allows us to compare the body measurements of individuals at the lower and upper extremes of the BMI distribution.

```python
import numpy as np
bmi = female[:,7]
valid_indices = ~np.isnan(bmi)
bmi_valid = bmi[valid_indices]
zfemale_valid = zfemale[valid_indices]
sorted_indices = np.argsort(bmi_valid)
lowest_bmi = zfemale_valid[sorted_indices[:5]]

highest_bmi = zfemale_valid[sorted_indices[-5:]]

print("Standardized measurements of 5 participants with LOWEST BMI:")
print(lowest_bmi)

print("\nStandardized measurements of 5 participants with HIGHEST BMI:")
print(highest_bmi)
```

## Interpretation
The printed rows represent the standardized body measurements of female participants with the lowest and highest BMI values.

Participants with the lowest BMI generally show lower standardized values for body measurements such as weight, waist circumference, and hip circumference. This indicates that these individuals tend to have smaller body sizes relative to the overall population.

In contrast, participants with the highest BMI tend to have higher standardized values for weight, waist circumference, and hip circumference. These measurements indicate larger body size and greater body mass relative to the population average.

This comparison highlights how BMI is strongly associated with other body measurements, especially weight and waist circumference. Individuals with higher BMI typically have higher values in these related measurements.
